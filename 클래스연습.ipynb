```python
# 포켓몬

import random

class Pokemon():
    def __init__(self, name, hp, level=10):
        self.name = name
        self.level = level
        self.hp = hp

    def attack(self, opponent):
        self.opponent = opponent
        damage = self.level + 3
        print(f'{self.name}가 {opponent.name}에게 {damage}의 데미지를 입혔습니다!')
        opponent.hp -= damage

    def is_knocked_down(self):
        return self.hp <= 0

    def status(self):
        print(f'{self.name}의 체력은 {self.hp}이 남아있습니다...')

pikachu = Pokemon('피카츄', 100, level = 10)
raichu = Pokemon('라이츄', 100, level = 10)

print('앗 야생의 라이츄가 나타났다!')
while True:
    # 피카츄가 라이츄를 공격한다.
    pikachu.attack(raichu)
    if raichu.is_knocked_down():
        print(f'{raichu.name}는 쓰러졌다! {pikachu.name}는 승리했다!')
        break

    # 라이츄가 피카츄를 공격한다.
    raichu.attack(pikachu)
    if pikachu.is_knocked_down():
        print(f'{pikachu.name}는 쓰러졌다! {raichu.name}는 승리했다!')
        break

    pikachu.status()
    raichu.status()
    print('-----------')


    

        
    
        
```

```python
# 러시안 룰렛 정답...

import random

class Shooter:
    def __init__(self, name, hp, accuracy, ammo):
        self.name = name  # 총잡이의 이름
        self.hp = hp  # 총잡이의 체력
        self.accuracy = accuracy  # 총잡이의 정확도 (0~100)
        self.ammo = ammo  # 총잡이가 가진 총알 수

    def attack(self, opponent):
        if self.ammo > 0:  # 총알이 남아있는 경우에만 공격
            # 공격 성공 여부 결정: 정확도에 따라 랜덤으로 성공/실패
            hit_chance = random.randint(1, 100)
            if hit_chance <= self.accuracy:
                # 공격 성공: 상대방 체력 감소
                damage = random.randint(10, 30)  # 공격 데미지는 10~30 사이의 랜덤 값
                print(f'{self.name}가 {opponent.name}에게 {damage}의 데미지를 입혔습니다!')
                opponent.hp -= damage
            else:
                print(f'{self.name}의 공격이 빗나갔습니다!')
            # 총알 소모
            self.ammo -= 1
        else:
            print(f'{self.name}는 총알이 없습니다!')

    def is_knocked_down(self):
        return self.hp <= 0  # 체력이 0 이하일 경우 쓰러짐

    def status(self):
        print(f'{self.name}의 체력: {self.hp}, 총알: {self.ammo}, 정확도: {self.accuracy}%')

def game(shooter1, shooter2):
    print(f'{shooter1.name}와 {shooter2.name}의 대결이 시작됩니다!')
    
    # 경기는 번갈아 공격하며 진행
    while shooter1.ammo > 0 and shooter2.ammo > 0:
        # 플레이어1이 공격
        shooter1.attack(shooter2)
        if shooter2.is_knocked_down():
            print(f'{shooter2.name}는 쓰러졌습니다! {shooter1.name}가 승리했습니다!')
            break

        # 플레이어2가 공격
        shooter2.attack(shooter1)
        if shooter1.is_knocked_down():
            print(f'{shooter1.name}는 쓰러졌습니다! {shooter2.name}가 승리했습니다!')
            break
        
        # 각 총잡이의 상태 출력
        shooter1.status()
        shooter2.status()
        print('-----------')

    # 총알이 다 떨어지면 게임 종료
    if shooter1.ammo == 0 and shooter2.ammo == 0:
        print('양쪽 모두 총알이 다 떨어졌습니다. 경기가 종료되었습니다.')
        if shooter1.hp > shooter2.hp:
            print(f'{shooter1.name}가 승리했습니다!')
        elif shooter2.hp > shooter1.hp:
            print(f'{shooter2.name}가 승리했습니다!')
        else:
            print('무승부입니다!')

# 총잡이 객체 생성
shooter1 = Shooter('철수', 100, accuracy=80, ammo=5)
shooter2 = Shooter('영희', 100, accuracy=75, ammo=5)

# 경기를 진행
game(shooter1, shooter2)
```
